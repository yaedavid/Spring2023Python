# Задание 1

```python
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

# Выдаёт график исходной функции
def orig_func(n):
    x = np.linspace(1, 15, n)
    plt.grid()
    plt.title('Исходная функция')
    plt.plot(x, f(x))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def approx(n):
    dots = np.linspace(1, 15, n)

    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)

    w = np.linalg.solve(a, b)  # точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение
    
    return res, b


orig_func(250)

n = 15
res = approx(n)
plt.grid()
plt.title(f'Полиномиальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
orig_func(250)

n = 5
res = approx(n)
plt.grid()
plt.title(f'Полиномиальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
orig_func(250)

n = 100
res = approx(n)
plt.grid()
plt.title(f'Полиномиальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
print("Мы видим, что при увеличении числа n функция полиномиальной аппроксимации становится более \
гладкой и похожей на оригинальную функцию, тогда как при уменьшении числа n она становится ломаной \
и непохожей на оригинальную функцию.")
```

# Задание 2

```python
import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)

    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)
    for i in range(len(b)):
        b[i] += np.random.normal(0, 0.01)

    w = np.linalg.solve(a, b)  # точки в полиноме 
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение
    
    return res, b


orig_func(250)

n = 15
res = approx(n)
plt.grid()
plt.title(f'Полиномиальная аппроксимация с шумом при n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
orig_func(250)

n = 5
res = approx(n)
plt.grid()
plt.title(f'Полиномиальная аппроксимация с шумом при n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
orig_func(250)

n = 100
res = approx(n)
plt.grid()
plt.title(f'Полиномиальная аппроксимация с шумом при n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
print("Из графиков видно, что при введении шума точность полиномиальной аппроксимации ухудшается, при этом \
в большей степени она влияет на аппроксимации с малым n.")
```

# Задание 3

```python
N = 100
def least_squares(N, n):
    dots = np.linspace(1, 15, N)
    a = np.array([[x**j for j in range(n)] for x in dots])
    b = f(dots)
    for i in range(len(b)):
        b[i] += np.random.normal(0, 0.1)

    w, resid, rank, s = numpy.linalg.lstsq(a, b, rcond=None)  # точки в полиноме
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение 
    
    return res, b


orig_func(250)

n = 3
res = least_squares(N, n)
plt.grid()
plt.title(f'Метод наименьших квадратов при n = {n}')
plt.plot(np.linspace(1, 15, N), res[0])
plt.plot(np.linspace(1, 15, N), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
orig_func(250)

n = 6
res = least_squares(N, n)
plt.grid()
plt.title(f'Метод наименьших квадратов при n = {n}')
plt.plot(np.linspace(1, 15, N), res[0])
plt.plot(np.linspace(1, 15, N), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
print("Как можно увидеть, метод наименьших квадратов даёт не точный график при n = 3. При увеличении n график \
становится более точным.")
```
